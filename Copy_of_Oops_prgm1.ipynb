```python
# Base Class
class Member:
    def __init__(self, name, membershipID):
        self.name = name
        self.membershipID = membershipID

    def get_name(self):
        return self.name

    def get_membershipID(self):
        return self.membershipID

    def display(self):
        print("Name:", self.name)
        print("Membership ID:", self.membershipID)


# Derived Class: Regular Member
class Regular(Member):
    def __init__(self, name, membershipID):
        super().__init__(name, membershipID)

    def display(self):
        print("Membership Type: Regular")
        super().display()


# Derived Class: Premium Member
class Premium(Member):
    def __init__(self, name, membershipID):
        super().__init__(name, membershipID)

    def display(self):
        print("Membership Type: Premium")
        super().display()


# Object Creation
r1 = Regular("Prabhu", 101)
p1 = Premium("Kumar", 201)

print("----- Regular Member -----")
r1.display()

print("\n----- Premium Member -----")
p1.display()
```
