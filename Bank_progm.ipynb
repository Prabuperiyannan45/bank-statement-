```python
acc_pin_number = 1234
acc_holder_name = "aaa"
acc_balance = 20000

def deposit():
    global acc_balance
    amount = float(input("Enter deposit amount: "))
    acc_balance += amount
    print("Amount Deposited Successfully")
    print("Current Balance:", acc_balance)

def withdraw():
    global acc_balance
    amount = float(input("Enter withdrawal amount: "))
    if amount <= acc_balance:
        acc_balance -= amount
        print("Amount Withdrawn Successfully")
    else:
        print("Insufficient Balance")
    print("Current Balance:", acc_balance)

def check_balance():
    print("Current Balance:", acc_balance)

# Main Program
pin = int(input("Enter your pin: "))

if pin == acc_pin_number:
    print("Correct PIN")
    print("1. Deposit")
    print("2. Withdrawal")
    print("3. Balance Check")

    choice = int(input("What do you want to do: "))

    if choice == 1:
        deposit()
    elif choice == 2:
        withdraw()
    elif choice == 3:
        check_balance()
    else:
        print("Select correct option")
else:
    print("Wrong PIN")
```
